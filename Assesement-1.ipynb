### Store each customer's order details (customer name, product, price, category) as tuples inside a list

```python
order_details = [
    ("Ankit","Laptop", 200,"Electronics"),
    ("Priya","Mobile",150,"Electronics"),
    ("Komal","Vaccum cleaner",40,"Home"),
    ("Jayesh", "Gas stove", 30, "Home"),
    ("Ramesh", "Chair", 60, "Home"),
    ("Kirti", "Dress", 110, "Clothing"),
    ("Mahesh", "Shirt", 120, "Clothing"),
    ("Ankit","T-shirt",70,"Clothing"),
    ("Shyam","Laptop", 200,"Electronics"),
    ]
print(f"Order Details : {order_details}")
```

### Use a dictionary to map each product to its respective category

```python
product_category = {"Laptop": "Electronics","Dress": "Clothing","Mobile": "Electronics","Vaccum cleaner": "Home","Gas stove": "Home","Chair": "Home","Shirt":"Clothing"}
print(f"Product & Category : {product_category}")
```

### Create a set of unique product categories and Display all available product categories

```python
categories = set(product_category.values())
print(f"Product Categories {categories}:")
```

###  Use a loop to calculate the total amount each customer spends

### If the total purchase value is above $100, classify the customer as a high-value buyer

```python
customer_spending = {}                               # This creates an empty dictionary

for name,product,price,category in order_details:    # This loop goes through each tuple in the list order_details 
    if name not in customer_spending:                # Is this customer already in the dictionary?
        customer_spending[name] = 0                  # Now dictionary becomes:0
    customer_spending[name] += price                 # customer_spending[name] = customer_spending[name] + price

print(f"Customer Spending : {customer_spending}")


for customer, total in customer_spending.items():
    if total > 100:
        print(f"{customer} is a high-value buyer")
```

### If it is between $50 and $100, classify the customer as a moderate buyer and If it is below $50, classify them as a low-value buyer

```python
for customer, total in customer_spending.items():
    if total > 100:
        category = "High Value Buyer"
    elif total > 50:
        category = "Moderate Value Buyer"
    else:
        category = "Low Value Buyer"
        
    print(f"{customer} spends {total} = {category}")        
```

### Calculate the total revenue per product category and store it in a dictionary

```python
category_revenue = {}

for name,product,price,category in order_details:
    if category not in category_revenue:
        category_revenue[category] = 0
    category_revenue[category]  += price
print(f"Category Revenue :{category_revenue}")    
        
```

### Extract unique products from all orders using a set

```python
unique_products = set()

for name,product,price,category in order_details:
    unique_products.add(product)
print(f"Unique Products : {unique_products}")    
```

### Use a list comprehension to find all customers who purchased electronics

```python
electronics_customers = {name for name, product, price, category in order_details if category == "Electronics"}

print("Customers who bought Electronics:", electronics_customers)
```

### Identify the top three highest-spending customers using sorting

```python
top_customers = sorted(customer_spending.items(), key=lambda x: x[1], reverse=True)

top_three = top_customers[:3]

print(f"All Sorted customers based on price :{top_customers}")
print(f"Top 3 highest spending customers :{top_three}")
```

### Print a summary of each customer’s total spending and their classification


```python
for customer, total in customer_spending.items():

    if total > 100:
        category = "High-value buyer"

    elif total >= 50:
        category = "Moderate buyer"

    else:
        category = "Low-value buyer"

    print(f"{customer} spent ${total} and is a {category}")
```

### Use set operations to find customers who purchased from multiple categories

```python
customer_categories = {}

# Collect categories for each customer
for name, product, price, category in order_details:
    
    if name not in customer_categories:
        customer_categories[name] = set()
    
    customer_categories[name].add(category)

# Find customers with multiple categories
for customer, categories in customer_categories.items():
    
    if len(categories) > 1:
        print(f"{customer} purchased from multiple categories: {categories}")
```

### Identify common customers who bought both electronics and clothing

```python
electronics_customers = set()
clothing_customers = set()

for name, product, price, category in order_details:
    
    if category == "Electronics":
        electronics_customers.add(name)
        
    if category == "Clothing":
        clothing_customers.add(name)

common_customers = electronics_customers & clothing_customers

print("Customers who bought both Electronics and Clothing:", common_customers)
```

## Frequently Purchased Product

```python
product_count = {}

for name, product, price, category in order_details:
    
    if product not in product_count:
        product_count[product] = 0
        
    product_count[product] += 1

most_frequent = max(product_count, key=product_count.get)

print("Most frequently purchased product:", most_frequent)
```
